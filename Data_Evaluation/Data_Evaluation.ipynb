```python
import pandas as pd
import json

# Function to load JSON objects from a file
def load_json_objects(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            data.append(json.loads(line))
    return pd.DataFrame(data)

# Load the JSON files into pandas dataframes
receipts_df = load_json_objects('./receipts.json')
users_df = load_json_objects('./users.json')
brands_df = load_json_objects('./brands.json')


display(receipts_df)
```

```python
# Extracting and flattening rewardsReceiptItemList into a new DataFrame
receipt_items = []
for index, row in receipts_df.iterrows():
    receipt_id = row['_id']
    items = row['rewardsReceiptItemList']

    # Check if items is a list (handling NaN or missing values)
    if isinstance(items, list):
        for item in items:
            item['_id'] = receipt_id  # Adding receipt_id to each item
            receipt_items.append(item)
    else:
        pass

receipt_items_df = pd.DataFrame(receipt_items)

display(receipt_items_df.head())
```

```python
from datetime import datetime

users_df

# Parse the '_id' field to extract the string value if it is a dictionary
users_df['_id'] = users_df['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) else x)

# Parse the 'lastLogin' field to convert the timestamp to a datetime object if it is a dictionary
users_df['lastLogin'] = users_df['lastLogin'].apply(lambda x: datetime.fromtimestamp(x['$date'] / 1000) if isinstance(x, dict) else x)

# Parse the 'createdDate' field to convert the timestamp to a datetime object if it is a dictionary
users_df['createdDate'] = users_df['createdDate'].apply(lambda x: datetime.fromtimestamp(x['$date'] / 1000) if isinstance(x, dict) else x)
```
